```python
import numpy as np
import matplotlib.pyplot as plt

def Calcularcoeficientesmyb(x,y):
    n=np.size(x)
    promediox,promedioy=np.mean(x), np.mean(y)
    sumatoriaxy= np.sum((x-promediox)*(y-promedioy))
    sumatoriaxx= np.sum((x-promediox)**2)
    m=sumatoriaxy/sumatoriaxx
    b= promedioy-promediox*m 
    return(m,b)

def plot_regression(x,y,ecuacion):
    plt.scatter(x,y, color="g", marker="o", s=30)
    y_ecu=ecuacion[0]*x+ecuacion[1]
    plt.plot(x,y_ecu,color="b")
    cadena=str(f"Ecuaci√≥n de la recta: y =  {ecuacion[0]}x +  {ecuacion[1]}")
    plt.xlabel(cadena)
    plt.show()

def main():

    x=np.array([1,2,3,4,5])
    y=np.array([2,3,5,6,5])

    ecuacion=Calcularcoeficientesmyb(x,y)
    print("la pendiente = ", ecuacion[0], " y el intecepto con el eje y: ", ecuacion[1])
    plot_regression(x,y,ecuacion)

if __name__=="__main__":
    main()
```
